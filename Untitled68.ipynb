```python
def height_corr(x):
  feet, inches = x.split("-")
  feet = int(feet)
  inches = int(inches)
  converted_inches = feet*12
  new_height = converted_inches + inches
  return new_height
```

```python
df_nba["Height"] = df_nba["Height"].apply(lambda x : height_corr(x))
```
